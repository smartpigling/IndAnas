```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('data/douban_movie.txt')
df.head(1)
```

```python
#按类型查找
comedy_df = df[df['types'].apply(lambda x: True if '喜剧' in x else False)].drop_duplicates(['title'])
```

```python
#每天平均评价人数
from datetime import datetime 
df['release_date'] = pd.to_datetime(df['release_date'])
df['total_day'] = df['release_date'].map(lambda x: (datetime.now() - x).total_seconds()/ (3600 * 24))
df['daily_vote'] = df['vote_count'] / df['total_day']
df.sort_values('daily_vote', ascending=False)
```

```python
#电影评级
df['movie_level'] = pd.cut(df['score'], bins = [0, 7.0, 8.0, 9.0, float('Inf')], labels = ['C', 'B', 'A', 'S'])
df
```

```python
df.groupby('movie_level').size().reset_index(name='num')
```

```python
df.groupby('movie_level')['id'].nunique().reset_index(name='num')
```

```python
df.groupby(['movie_level', 'id']).agg({'score': np.mean, 'vote_count': np.max}).sort_values('score', ascending=False)
```
